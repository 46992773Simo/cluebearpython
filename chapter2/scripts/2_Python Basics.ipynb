### 2.5 面向对象的编程基本概念

#### 2.5.1 类的基本定义和使用

```python
class Hot_pot():
    pass
```

```python
class Hot_pot():
    cooking = "instant boiled"                  #烹饪方式属性
    season = "four seasons"                     #适用季节属性
    def cook(self):                            #定义cook函数，即方法
        print('涮火锅')   
    def eat(self,name):                         #定义eat函数
        self.name = name
        print(self.name,'是%s的烹饪方式,适合季节是%s。'%(self.cooking,self.season))  
```

#### 2.5.2 对象

```python
dish = Hot_pot()    #创建Hot_pot类的对象dish     
dish.cooking        #dish的cooking属性，返回值'instant boiled'
dish.season         #dish的season属性，返回值'four seasons'
dish.cook()         #dish的cook方法，返回值 涮火锅
dish.eat('四川火锅') #返回值 ''四川火锅 是instant boiled的烹饪方式,适合季节是four seasons。
```

```python
class Hot_pot():
    def __init__(self,cooking,season):
        self.cooking = cooking                      #属性
        self.season = season                        #属性
    def cook(self,):                            #定义cook函数，即方法
        print('涮火锅')   
    def eat(self,name):                         #定义eat函数
        self.name = name
        print(self.name,'是%s的烹饪方式,适合季节是%s。'%(self.cooking,self.season))
dish1 = Hot_pot('boiled','winter') 
```

```python
class Hot_pot_coupon:
   # 定义构造方法
    def __init__(self,name,mix):
        self.name=name
        self.mix=mix
    def introduce(self):
        print('%s的团购价：%d元'%(self.name,self.mix))
class New(Hot_pot_coupon):
    def __init__(self,name,mix,addition):
        #调用父类的构造方法
        Hot_pot_coupon.__init__(self,name,mix)
        self.addition=addition
    # 覆写父类的方法
    def introduce(self):
        print('%s的团购价：%d元,赠品是%s。'%(self.name,self.mix,self.addition))
M=Hot_pot_coupon('情人节情侣火锅套餐',70)
M.introduce() #返回值 情人节情侣火锅套餐的团购价：70元
P=New('情人节情侣火锅套餐',70,'巧克力')
P.introduce()
```

```python
class Hot_pot_coupon:
    def myMethod(self):
        print('调用父类方法')
class New(Hot_pot_coupon):
    def myMethod(self):
        print('调用子类方法')
c = New()
c.myMethod()
super(New, c).myMethod()
```
