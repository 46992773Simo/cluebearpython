```python
import pandas as pd 
import os 
```

# 3.1 初级篇   

```python
path = '../data'
raw_data = pd.read_excel(os.path.join(path, 'shops_nm.xlsx'), encoding='utf8', index_col=None)

print('数据的维度是：', raw_data.shape)
raw_data.head()
```

```python
# 判断是否存在完全一致的数据行
duplicated_data = raw_data[raw_data.duplicated() == True]
if len(duplicated_data) == 0:
    print('不存在完全一致的数据行')
else:
    print(duplicated_data)

# 判断是否存在重复的店名
duplicated_shops = raw_data['店名'][raw_data['店名'].duplicated() == True]
if len(duplicated_shops) == 0:
    print('不存在重复的店名')
else:
    print(duplicated_shops)

drop_duplicates = raw_data.drop_duplicates(subset=['店名'])
```

```python
is_null_data = raw_data[raw_data['评价数'].isnull() == True]
raw_data2 = raw_data[raw_data['评价数'].isnull() == False]
is_null_data
```

```python

filter_words = ['人均：', '人均', '大概', '左右', '差不多']
for i in range(len(raw_data2)):
    value = raw_data2['人均'].iloc[i]
    if type(value) is int or type(value) is float:
        continue
    for word in filter_words:
        if word in value:
            raw_data2.loc[i, '人均'] = raw_data2.loc[i, '人均'].replace(word, '')

raw_data2['人均'] = raw_data2['人均'].astype(float)
raw_data2.head() 
```

```python
raw_data2.describe()
```

# 3.2 进阶篇

```python
def clean_price(x):
    filter_words = ['人均：', '人均', '大概', '左右', '差不多']
    if type(x) is int or type(x) is float:
        return x 
    for word in filter_words:
        if word in value:
            x = x.replace(word, '')
    return x 

raw_data2['人均'] = raw_data2['人均'].apply(clean_price)
raw_data2['人均'] = raw_data2['人均'].astype(float)
raw_data2.head()
```

```python
def cate_shops(x):
    if x <= 2:
        return 0
    if x > 2 and x <= 3.5:
        return 1
    if x > 3.5:
        return 2
    
raw_data2['商家等级'] = raw_data2['评分'].apply(cate_shops)
raw_data2.groupby(by=['商家等级']).agg({'人均': 'mean'})
```

```python
coupon_data = pd.read_excel(os.path.join(path, 'coupon_nm.xlsx'), encoding='utf8', index_col=None)

coupon_data = coupon_data[coupon_data['团购评价'] != '暂无评价'][coupon_data['评价人数'] != '暂无评价']
coupon_data = coupon_data[coupon_data['购买人数'] != 0]
coupon_data = coupon_data.dropna()
coupon_data.head(2)
```

```python
print(type(coupon_data['到期时间'].iloc[0]))
coupon_data['到期时间'] = pd.to_datetime(coupon_data['到期时间'], format='%Y-%m-%d')
print(type(coupon_data['到期时间'].iloc[0]))

from datetime import datetime
filter_coupon_data = coupon_data[coupon_data['到期时间'] > datetime.strptime('2018-09-01', '%Y-%m-%d')]
filter_coupon_data.head(2)
```

```python
filter_coupon_data2 = filter_coupon_data.groupby('店名', group_keys=False).agg({
    '团购价': 'mean',
    '购买人数': 'mean',
})

filter_coupon_data2.head(10)
```

```python
filter_coupon_data2['店名'] = filter_coupon_data2.index
filter_coupon_data2.index = [i for i in range(len(filter_coupon_data2))]
filter_coupon_data2.head() 

merge_data = pd.merge(left=raw_data2, right=filter_coupon_data2, on='店名', how='inner')
merge_data['购买人数'] = merge_data['购买人数'].astype(int)
print(merge_data.shape)

merge_data.head()
```

```python
merge_data.to_excel(os.path.join(path, 'merge_shop_coupon_nm.xlsx'), encoding='utf8', index=False)
```
