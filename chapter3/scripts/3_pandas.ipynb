# 数据科学实践|Pandas篇

# 3.1 初级篇   

```python
import os 
import pandas as pd # 导入Pandas模块
path = "../data"
raw_data = pd.read_excel(os.path.join(path, "shops_nm.xlsx")) #读入数据
print('数据的维度是：', raw_data.shape)
raw_data.head() #查看数据的前5行
```

```python
# 判断是否存在完全一致的数据行
duplicated_data = raw_data[raw_data.duplicated() == True]
if len(duplicated_data) == 0:
    print('不存在完全一致的数据行')
else:
    print(duplicated_data)
```

```python
print(raw_data.iloc[0,:])
raw_data.iloc[1,:] = raw_data.iloc[0,:]
```

```python
# 判断是否存在完全一致的数据行
duplicated_data = raw_data[raw_data.duplicated() == True]
if len(duplicated_data) == 0:
    print('不存在完全一致的数据行')
else:
    print(duplicated_data)
```

```python
# 判断是否存在重复的店名
duplicated_shops = raw_data['店名'][raw_data['店名'].duplicated() == True]
if len(duplicated_shops) == 0:
    print('不存在重复的店名')
else:
    print(duplicated_shops)
```

```python
drop_duplicates_shops = raw_data.drop_duplicates(subset=['店名'])
drop_duplicates_shops.head()
```

```python
print(raw_data[raw_data['评分']==5])
```

```python
is_null_data = raw_data[raw_data['评价数'].isnull() == True]
raw_data2 = raw_data[raw_data['评价数'].isnull() == False]
is_null_data
```

```python
s1_average = raw_data2.loc[1,'人均']
print(type(s1_average))

s2_average = raw_data2.loc[2,'人均']
print(type(s2_average))
```

```python
filter_words = ['人均：', '人均', '大概', '左右', '差不多'] #定义需要过滤的词
for i in range(len(raw_data2)):
    value = raw_data2['人均'].iloc[i] #取出人均这一行中的值
    if type(value) is int or type(value) is float: 
        continue #判断该值是否为整数或者浮点类型数字，如果满足则跳过进入下一步
    for word in filter_words:
        if word in value: #判断需要过滤的词是否在value中，如果在则去除
            raw_data2.loc[i, '人均'] = raw_data2.loc[i, '人均'].replace(word, '')

raw_data2.head() 
```

```python
raw_data2.describe()
```

# 3.2 进阶篇

```python
def clean_price(x):
    filter_words = ['人均：', '人均', '大概', '左右', '差不多']
    if type(x) is int or type(x) is float:
        return x 
    for word in filter_words:
        if word in value:
            x = x.replace(word, '')
    return x 

raw_data2['人均'] = raw_data2['人均'].apply(clean_price)
raw_data2['人均'] = raw_data2['人均'].astype(float)
raw_data2.head()
```

```python
def cate_shops(x):
    if x <= 2:
        return 0
    if x > 2 and x <= 3.5:
        return 1
    if x > 3.5:
        return 2
    

raw_data2['商家等级'] = raw_data2['评分'].apply(cate_shops) #构建商家等级
raw_data2.groupby(by=['商家等级']).agg({'人均': 'mean'}) #根据商家等级分组，然后对于人均按照均值进行整合
```

```python
coupon_data = pd.read_excel(os.path.join(path, "coupon_nm.xlsx")) #读入数据

#清洗数据，把没有评价的团购，购买人数为0的团购等去掉
coupon_data = coupon_data[coupon_data['团购评价'] != '暂无评价'][coupon_data['评价人数'] != '暂无评价']
coupon_data = coupon_data[coupon_data['购买人数'] != 0]
coupon_data = coupon_data.dropna()
coupon_data.head(2)
```

```python
print(type(coupon_data['到期时间'].iloc[0])) #打印第一个团购的到期时间的数据类型
coupon_data['到期时间'] = pd.to_datetime(coupon_data['到期时间'], format='%Y-%m-%d') #转换团购到期时间为时间类型数据
print(type(coupon_data['到期时间'].iloc[0])) #重新查看第一个团购的到期时间的数据类型

from datetime import datetime # 字符串转化为时间类型
filter_coupon_data = coupon_data[coupon_data['到期时间'] > datetime.strptime('2018-09-01', '%Y-%m-%d')] #获得2018年9月1日后的团购
filter_coupon_data.head()
```

```python
filter_coupon_data2 = filter_coupon_data.groupby('店名').agg({
    '团购价': 'mean',
    '购买人数': 'mean',
}) #按照店名分拆，用团购价与购买人数均值进行聚合
filter_coupon_data2.head(10)
```

```python
filter_coupon_data2['店名'] = filter_coupon_data2.index #添加新列’店名‘
filter_coupon_data2.index =  list(range(len(filter_coupon_data2))) #重新按照编号构建Index
filter_coupon_data2.head(10) 
```

```python
merge_data = pd.merge(left=raw_data2, right=filter_coupon_data2, on='店名', how='inner') #按照店名合并
merge_data['购买人数'] = merge_data['购买人数'].astype(int) #转化为整数
print(merge_data.shape)

merge_data.head()
```

```python
merge_data.to_excel(os.path.join(path, 'merge_shop_coupon_nm.xlsx'), encoding='utf8', index=False)
```
