### 1利用open函数读文件

```python
# 打开一个文件
f = open("food.txt", "w")
# 在文件中写入两行话
f.write("西安火锅真棒。\n我喜欢西安的火锅!!\n")
# 关闭打开的文件
f.close()
```

```python
##读入名为food的文件
f = open("food.txt", "r")
##读入4个字符长度
str = f.read(4)
print(str)
##关闭文件
f.close()
```

```python
f = open("food.txt", "r")
str = f.readline()
print(str)
f.close()
```

```python
f = open("food.txt", "r")
a = f.readlines()
print(a)
f.close()
```

```python
file = open('food.txt', 'r')
for line in file:
    print (line)
```
