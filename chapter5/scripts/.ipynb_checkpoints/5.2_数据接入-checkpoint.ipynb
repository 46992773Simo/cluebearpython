# 5.2 数据接入

## 2 外部数据接入

#### 读取数据

```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import os 
```

```python
path = '../data'    #定义路径
merge_data = pd.read_excel(os.path.join(path, 'merge_shop_coupon_nm.xlsx'), encoding='utf8', index=False)    #读取数据
merge_data.head()    #查看数据
```

#### 筛选数据

```python
model_cols = ['城市', '评分', '评价数', '人均', '商家等级', '购买人数', '团购价']    #定义模型数据列
model_data = merge_data.loc[:, model_cols]    #筛选数据
```

#### DataFrame与NumPy数组的互相转化

```python
num_cols = ['评分', '评价数', '人均', '商家等级', '购买人数', '团购价']    #定义数值变量所在列
```

```python
model_data.loc[:, num_cols].values    #将数值变量转化为numpy数组
```

```python
model_data.values    #将所有建模数据都转化为numpy数组（不均匀）
```

```python
df = pd.DataFrame(model_data.values, columns = ['城市', '评分', '评价数', '人均', '商家等级', '购买人数', '团购价'])    #数组转化为DataFrame
df    #输出df查看结果
```
